# store_financials_v1

```python
store_financials = load_csv(DATA_PATH / "store_financials_v1.csv")
```

```python
store_financials.shape
store_financials.info()
store_financials.isna().mean().sort_values(ascending=False)
```

```python
# check for duplicates
store_financials[store_financials.duplicated('codstr', keep= False)]
```

```python
# negative costs found
store_financials[store_financials['total_costs_r12m'] < 0]
```

```python
# Mark tenants who have received the rent incentive
store_financials['is_subsidized'] = (store_financials['total_costs_r12m'] < 0).astype(int)
```

```python
# unite the currency
store_financials['cur_code'].value_counts()
```

```python
exchange_rates_to_eur = {
    'EUR': 1.0,        
    'USD': 1.0884,
    'PLN': 4.2637, 
    'GBP': 0.8406,
    'CZK': 25.112,
    'SEK': 11.2874,
    'DKK': 7.4602
}

store_financials['sales_r12m'] = pd.to_numeric(store_financials['sales_r12m'], errors='coerce')
store_financials['total_costs_r12m'] = pd.to_numeric(store_financials['total_costs_r12m'], errors='coerce')

store_financials['fx_rate'] = store_financials['cur_code'].map(exchange_rates_to_eur)
store_financials['sales_eur'] = store_financials['sales_r12m'] / store_financials['fx_rate']
store_financials['costs_eur'] = store_financials['total_costs_r12m'] / store_financials['fx_rate']
```

```python
store_financials.to_csv('store_financials.csv')
```
